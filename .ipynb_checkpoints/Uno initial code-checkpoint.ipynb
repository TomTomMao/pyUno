```python
from random import shuffle

##Card Creation

## Cards have 2 features, colours and function
## To generate cards, we need to create lists of all possible colours and functions
## Normal Colours and Red, Yellow, Blue, Green
## Normal Funtions are numbers 0-10, skip turn, reverse order, draw 2
## Special Colour is Black, matches all normal colours
## Special Functions are wildcard and draw 4, both allowing to pick the colour of play to follow
## To aid in game logic, these will be segmented into multiple variables to be checked against eachother
## Once a variables are created for each card type and colour, these will be conbined to generate "allcards" containing all possible cards

numbers = list(range(1,10)) + list(range(10))
special = ["skip_turn","reverse","draw_2"] *2
black_special = ["wildcard", "draw_4"] *4
allfunctions = numbers + special + black_special
normalcolours = ["red","yellow","green","blue"]
allcolours = normalcolours + ["black"]

allcards = []

for number in numbers:
    for colour in normalcolours:
        allcards.append([colour,number])
for function in special:
    for colour in normalcolours:
        allcards.append([colour,function])
for function in black_special:
    allcards.append(["black",function])

    
## Each card will be a object within object class

class card:
    def __init__(self,colour,function):
        self.colour = colour
        self.function = function

    ## Each card will also be able to check itself against another, to check if the card is valid to be played
    
    def valid_play(self,top_card):
        return(self.colour == top_card.colour or
               self.function == top_card.function or
               self.colour == "black")
        
        
## Information on each player will also be stored in a class

        
class player:

    def __init__(self,hand,player_id):
        self.player_id = player_id
        self.hand = [hand]
        
class computerplayer(player):
    pass
    
class humanplayer(player):
    pass
## creation of a class used to store functions of the game

class game:
##define init to contain all variables that pay be needed in the game
    def __init__(self):
        self.deck = self.createdeck()
        self.discard = []
        self.topcard = self.initialtopcard()
        self.players = [player(self.starthand(),n) for n in range(8)]
        self.playercycle = []
        self.currentplayer = []
    
## at the start of game, a deck of uno cards will need to be created
    def createdeck(self):
        deck = [card(colour,function) for colour , function in allcards]
        shuffle(deck)
        return(deck)
        
##start of game, draw 7 cards
    def starthand(self):
        return([self.deck.pop() for n in range(7)])
        
        
        
##players will need to be able to draw cards throughout the game
    def drawcard(self,playerid):
        carddrawn = self.deck.pop(1)
        self.players[playerid].hand[0].append(carddrawn)
        
##to enable swapping of cards, cards must be able to be discarded from hand
##card is the cards index within the players hand
##adds card to same pile as playing a card, but does not change the "top card" value
    def discardcard(self,playerid,card):
        self.discard.append(self.players[playerid].hand[0].pop(card))
        
##function to swap a card
##discards selected card by index, replaces it by drawingcard
    def swapcard(self,playerid,card):
        self.discardcard(playerid,card)
        self.drawcard(playerid)
        
## on game initialising, a top card will need to be drawn
## this top card moved into the discard pile

    def initialtopcard(self):
        return([self.deck.pop(1)])
    
## playing a card involves the following steps
## check if card is valid to play, if not, return false
## if playable, put card in discard, and replace top card with played card

    def playcard(self,playerid,card):
        pass
    
## a player wins if they have 0 cards
    def checkwin(self,playerid):
        pass
        
        
```
